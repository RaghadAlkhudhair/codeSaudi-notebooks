### Example: The Runner Race

````python
for i in range(5):
    print("Runner " + str(i) + " is fast!")
````

Make a loop that prints "Clap!" 10 times!

````python
# Make 10 claps:
for i in range(10):
    print("Clap! üëè (#" + str(i + 1) + ")")
````

### üåü Bonus: Loop Through a List!
Print every friend's name with a greeting!

````python
friends = ["Ali", "Sara", "Omar", "Lina"]

for friend in friends:
    print("Hello, " + friend + "! Welcome to the party! üéâ")

print("\nTotal friends:", len(friends))
````
