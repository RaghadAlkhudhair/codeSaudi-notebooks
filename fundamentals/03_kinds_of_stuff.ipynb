In Python, every piece of data has a TYPE â€” like sorting toys into piles!

- **STRING** = Text (words in quotes)
- **INTEGER** = Whole numbers
- **BOOLEAN** = True or False

---

### Example: The 3 Types

```python
name = "Alex"      # STRING (Text)
age = 7             # INTEGER (Number)
hungry = True       # BOOLEAN (Yes/No)

print(type(name))   # <class 'str'>
print(type(age))    # <class 'int'>
print(type(hungry)) # <class 'bool'>
```
