### üéØ Your Turn!
You have 10 apples. You give 3 to a friend and 2 to your teacher. How many do you have left?

```python
apples = 10
give_friend = 3
give_teacher = 2

# Calculate how many are left:
left = apples - give_friend - give_teacher
print("Apples left:", left)
```

### üåü Bonus: Build a Score Calculator!
A game gives you 10 points per coin and 50 points per star. Calculate your total score!

```python
coins = 7
stars = 3
points_per_coin = 10
points_per_star = 50

total_score = (coins * points_per_coin) + (stars * points_per_star)
print("Your total score is:", total_score, "points! üéÆ")
```
