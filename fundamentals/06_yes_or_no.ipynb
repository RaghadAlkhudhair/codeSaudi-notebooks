Python can answer Yes/No questions using **True** and **False**!

- `==` Is it the same?
- `!=` Is it different?
- `>` Is it bigger?
- `<` Is it smaller?

---

### Example: Asking Python Questions

```python
print(10 == 10)   # Is 10 equal to 10? True!
print(5 > 100)    # Is 5 bigger than 100? False!
```
