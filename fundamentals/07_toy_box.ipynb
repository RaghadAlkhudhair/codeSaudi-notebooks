### ðŸŒŸ Bonus: Add & Remove!
Use `.append()` to add a new toy and `.remove()` to take one out!

```python
toys = ["Car", "Robot", "Ball"]
print("Before:", toys)

# Add a new toy:
toys.append("Doll")
print("After adding:", toys)

# Remove a toy:
toys.remove("Car")
print("After removing:", toys)

# How many toys are left?
print("Total toys:", len(toys))
```
